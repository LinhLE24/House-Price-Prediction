# Training the Algorithm

```python

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)

print ('Coefficients: ', regressor.intercept_)

print ('Coefficients: ', regressor.coef_)
```

# Model training and evaluation

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.metrics import r2_score, explained_variance_score

def Performance_metrics(y_pred, y_test):
    y_pred = y_pred.squeeze()
#     y_pred cua test ( validation) data
    y_test = y_test.squeeze()

    print("Mean square error (MSE): %.2f" % np.mean((y_pred - y_test) ** 2))
    print("Root mean square error (RMSE): %.2f" % np.sqrt(np.mean((y_pred - y_test) ** 2)))
    print("Mean absolute error (MAE): %.2f" % np.mean(abs(y_pred - y_test)))
    print("Coefficient of determination (R^2): %.2f" % r2_score(y_test, y_pred))
```

```python
y_pred=regressor.predict(X_test)
Performance_metrics(y_pred, y_test)
```

With r-squared of 0.80 reveals that 80% of the data fit the regression model. Generally, If we get a higher r-squared indicates a better fit for the model.
